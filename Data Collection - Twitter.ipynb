```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
location = "12.19400,123.37222,802.57km"

query_location = f'(sinovac OR astrazeneca OR moderna OR pfizer OR "covid vaccine" OR #COVIDVaccinePH OR #VaccinePH OR #GetVaccinatedPH OR #VaccineForAllPH OR #VaccineRolloutPH OR #VaccineSafetyPH OR #VaccineEfficiencyPH) until:2023-03-31 since:2020-01-01 geocode:"{location}"'
query_near = f'(sinovac OR astrazeneca OR moderna OR pfizer OR "covid vaccine" OR #COVIDVaccinePH OR #VaccinePH OR #GetVaccinatedPH OR #VaccineForAllPH OR #VaccineRolloutPH OR #VaccineSafetyPH OR #VaccineEfficiencyPH) until:2023-03-31 since:2020-01-01 near:"Masbate" within:803km'
```

```python
result = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(query_location).get_items(), 10000))
result_2 = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(query_near).get_items(), 10000))
```

```python
result.shape, result_2.shape
```

```python
# Combine the two dataframes
results = pd.concat([result, result_2], ignore_index=True)

# Drop duplicates
results.drop_duplicates(subset=['id'], inplace=True)
results.shape
```

```python
# reset index
results.reset_index(drop=True, inplace=True)
results.head()
```

```python
# Save to csv
results.to_csv('twitter_data.csv', index=False)
```

```python
# Load csv
results = pd.read_csv('twitter_data.csv')
```

```python
# only keep these columns: 'id', 'url', 'date', 'rawContent', 'renderedContent'
tweet_data = results[['id', 'url', 'date', 'renderedContent']]
tweet_data.head()
```

```python
import preprocessor as p
```
